```python
df.FlightStatus.value_counts(normalize=True)
```

0 = flights that arrived on time = 33.72%

Meaning 1 of 3 flights is normally delayed 

```python
# Create a new dataframe with only the columns of interest: 'Unique carrier' and 'Flight status'

dfPFD = df[['UniqueCarrier', 'FlightStatus']]
dfPFD.head()
```

```python
# 2. Shape to check the size of the dataframe and that we have not lost any rows for any reason
dfPFD.shape
```

```python
# 3. Non-normalize value_count()
dfPFD.FlightStatus.value_counts()
```

```python
# 4. Create another dataframe (test_1) grouping the airline names and adding all the FLIGHT_STATUS values
# This way you will obtain all the delayed flights
test1 = dfPFD.groupby(['UniqueCarrier']).sum().reset_index()

# 5. Create another dataframe (test_2) from the value_counts directly so that it has all the flights added per airline
# Rename the axis to OP_CARRIER and reset_index to TOTAL_FLIGHTS
test2 = df.UniqueCarrier.value_counts().rename_axis('UniqueCarrier').reset_index(name='TotalFlights')

# 6. Merge the two dataframes (test_1 and test_2) using the OP_CARRIER as the common column
df_fp = pd.merge(test1, test2, on='UniqueCarrier')

# 7. Create a new column that will represent the percentage of delay flights
df_fp['PercentageDelayed'] = round((df_fp['FlightStatus']*100)/df_fp['TotalFlights'], 2)
df_fp.rename({'TotalFlights': 'DelayedFlights'}, axis=1, inplace=True)

# 8. Display the entire dataframe
df_fp.head(30)
```

##### The flights delayed, total number of flight and Percentages are arranged Alphabetically 

```python
# 9. Double checking that that total delayed flights match from both sources
SUM = df_fp.DelayedFlights.sum()
print('Total number of delayed flights:', SUM)
```

```python
# 10. Plot with the 33.72% threshold value as a horizontal red line
plt.figure(figsize=(20, 10))
df_fp.groupby('UniqueCarrier').PercentageDelayed.sum().sort_values(ascending=False).plot.bar(fontsize=14)
plt.hlines(y=37.52, xmin=-1, xmax=18, colors='r', linestyles='solid', label='cutoff')
plt.title('PERCENTAGE OF DELAYED FLIGHTS BY AIRLINE', fontsize=20)
plt.xlabel('Airline', fontsize=16)
plt.ylabel('Percentage', fontsize=16)
plt.rc('xtick',labelsize=14)
plt.rc('ytick',labelsize=14)
plt.show()
```

### Average Time Delay per Airline


```python
df.head(10)
```

```python
df = df.drop(['Unnamed: 0'], axis=1)
```

```python
df.head(10)
```

```python
Test3 = df[['UniqueCarrier', 'ArrDelay']]
Test3 = Test3.groupby(['UniqueCarrier']).mean().reset_index()
Test3.rename({'ArrDelay': 'AverageDelay'}, axis=1, inplace=True)
Test3 = Test3.sort_values(['AverageDelay']).reset_index(drop=True)
Test3.head(20)
```

```python
Test4 = df[['UniqueCarrier', 'DepDelay']]
Test4 = Test4.groupby(['UniqueCarrier']).mean().reset_index()
Test4.rename({'DepDelay': 'AverageDelay'}, axis=1, inplace=True)
Test4 = Test4.sort_values(['AverageDelay']).reset_index(drop=True)
Test4.head(13)
```

```python
fig = plt.figure(1, figsize=(15,9))
sns.set(font_scale=1.6)
ax = sns.barplot(x='AverageDelay', y='UniqueCarrier', data=Test3)
ax.set_title('AVERAGE DELAY ON ARRIVAL BY AIRLINES (MINS)', fontsize=18)
ax.set_xlabel('AVERAGE DELAY BY AIRLINE (MINS)', fontsize=14)
plt.show()
```

```python
fig = plt.figure(1, figsize=(15,9))
sns.set(font_scale=1.6)
ax = sns.barplot(x='AverageDelay', y='UniqueCarrier', data=Test4)
ax.set_title('AVERAGE DELAY ON DEPARTURE BY AIRLINES (MINS)', fontsize=18)
ax.set_xlabel('AVERAGE DELAY BY AIRLINE (MINS)', fontsize=14)
plt.show()
```

1. We know that UA, DL, AA have the most number of flights respectively. But averagely AA has the least delays in terms of minutes between 3-5mins delays for both arrival and departure 

2. DL being the 2nd airline with the most number of flights has the on average a 9mins delay on departure and 7mins delay on arrival. 

3. UA on average delays for 8mins on both departure & arrival

#### Popular Destinations with Average Arrival Delays 

```python
Test5 = df[['Dest', 'ArrDelay']]
Test5 = Test5.groupby(['Dest']).mean().reset_index()
Test5.head(10)
```

```python
# Number of destinations
Test5.shape
```

```python
Test_5= df.Dest.value_counts().rename_axis('Dest').reset_index(name='FLIGHTS').iloc[0:30]
Test_5.head(30)
```

```python
Test5_5 = pd.merge(Test_5, Test_5, on='Dest')
Test5_5.head()
```

```python
Test5_5 = Test5_5.set_index('Dest')

fig = plt.figure(figsize=(15,7)) # Create matplotlib figure

ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as a
width = .3

Test5_5.FLIGHTS.plot(kind='bar',color='grey',ax=ax, width=width, position=0)
Test5_5.ArrDelay.plot(kind='bar',color='red', ax=ax2, width=width, position=1)

plt.hlines(y=4.90, xmin=-1, xmax=30, colors='b', linestyles='dashed', label='cutoff')

#ax.grid(None, axis=1)
ax2.grid(None)

ax.set_title('MOST POPULAR DESTINATIONS vs AVERAGE ARRIVAL DELAY', fontsize=18)
ax.set_ylabel('NUMBER OF FLIGHTS', fontsize=16)
ax2.set_ylabel('AVERAGE ARRIVAL DELAY (mins)')
ax.set_xlabel('DESTINATION CITIES', fontsize=16)

plt.show()
```

So horizontal dotted line across the graph shows the average delay threshold on arrival for the top 30 destinations is 5mins. ORD & ATL have one of the hughest number of landing and an average delay of 5mins. San Francisco on the other has a highest minute average delay of nearly 20mins so is Philadelphia, San-Diego and Seattle-Tacoma International Airport
