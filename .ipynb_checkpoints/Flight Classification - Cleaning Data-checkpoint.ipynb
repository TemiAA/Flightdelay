```python
df = pd.read_csv('year2000(1).csv')
df.head()
```

```python
df = df.drop(['Cancelled','Diverted'], axis=1)
```

```python
df.isna().any()[lambda x: x]
```

```python
# Total number of missing values
df.isnull().sum().sum()
# Replace missing values with a number
df['DepTime'].fillna(0, inplace=True)
df['ArrTime'].fillna(0, inplace=True)
df['ActualElapsedTime'].fillna(0, inplace=True)
df['ArrDelay'].fillna(0, inplace=True)
df['DepDelay'].fillna(0, inplace=True)
df['Distance'].fillna(0, inplace=True)
df.head(10)
```
